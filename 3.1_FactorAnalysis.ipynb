```python
import pandas as pd
from sklearn.decomposition import FactorAnalysis
import matplotlib.pyplot as plt
```

```python
# read imputed data
X = pd.read_csv('selectedX.csv',index_col=0)
X.shape
```

```python
# factor analysis
fa = FactorAnalysis(n_components=2)
Xt = fa.fit_transform(X)
```

```python
# plot and save
plt.plot(Xt[:,0],Xt[:,1],'.')
pd.DataFrame(Xt,columns=['FA1','FA2']).to_csv('FAresult.csv',index=False)
```
